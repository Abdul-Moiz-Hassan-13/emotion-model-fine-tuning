```python
import pandas as pd
import os

train = pd.read_csv(os.path.join(path, "train.tsv"), sep="\t", header=None)
val   = pd.read_csv(os.path.join(path, "dev.tsv"), sep="\t", header=None)
test  = pd.read_csv(os.path.join(path, "test.tsv"), sep="\t", header=None)

train.columns = ["text", "label", "id"]
val.columns   = ["text", "label", "id"]
test.columns  = ["text", "label", "id"]

print("✅ Reloaded with correct headers!")
print(train.head())

```

```python
print("Train shape:", train.shape)
print("Unique labels:", train['label'].nunique())
print(train.head(3))
```

```python
emotions_path = os.path.join(path, "emotions.txt")

with open(emotions_path, "r") as f:
    emotions = [line.strip() for line in f.readlines()]

print("✅ Total emotions:", len(emotions))
print(emotions)
```

```python
id2label = {i: emotion for i, emotion in enumerate(emotions)}
label2id = {emotion: i for i, emotion in enumerate(emotions)}

train["emotion"] = train["label"].map(id2label)
val["emotion"]   = val["label"].map(id2label)
test["emotion"]  = test["label"].map(id2label)

print(train[["text", "label", "emotion"]].head())
```

```python
for i, emotion in enumerate(emotions):
    print(f"{i:2d} → {emotion}")
```

```python
def simplify_label(label):
    return int(label.split(",")[0])

train["label"] = train["label"].apply(simplify_label)
val["label"]   = val["label"].apply(simplify_label)
test["label"]  = test["label"].apply(simplify_label)

train["emotion"] = train["label"].map(id2label)
val["emotion"]   = val["label"].map(id2label)
test["emotion"]  = test["label"].map(id2label)

print(train[["text", "label", "emotion"]].head())


```

```python
from datasets import Dataset, DatasetDict
from transformers import AutoTokenizer

train_ds = Dataset.from_pandas(train)
val_ds   = Dataset.from_pandas(val)
test_ds  = Dataset.from_pandas(test)

dataset = DatasetDict({
    "train": train_ds,
    "validation": val_ds,
    "test": test_ds
})

tokenizer = AutoTokenizer.from_pretrained("distilroberta-base")

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=128)

tokenized_datasets = dataset.map(tokenize, batched=True)

print(tokenized_datasets)
```

```python
from transformers import AutoModelForSequenceClassification

num_labels = len(id2label)
model_name = "distilroberta-base"

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id
)

print("✅ Model initialized for fine-tuning!")
```

```python
from transformers import TrainingArguments, Trainer
import evaluate
import numpy as np

accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=predictions, references=labels)["accuracy"],
        "f1": f1.compute(predictions=predictions, references=labels, average="weighted")["f1"]
    }

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",    
    save_strategy="epoch",    
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    report_to=[],            
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

print("✅ Trainer initialized and ready for fine-tuning!")
```

```python
train_result = trainer.train()

trainer.save_model("./emotion_model")
tokenizer.save_pretrained("./emotion_model")

print("✅ Fine-tuning complete! Model saved to ./emotion_model")
```

```python
test_metrics = trainer.evaluate(tokenized_datasets["test"])
print("✅ Test results:", test_metrics)
```

```python
from transformers import pipeline

emotion_pipeline = pipeline(
    "text-classification",
    model="./emotion_model",
    tokenizer="./emotion_model",
    return_all_scores=True
)

examples = [
    "I can’t believe this happened, I’m so upset!",
    "Wow, that was such a beautiful surprise!",
    "I’m really nervous about tomorrow.",
    "Finally done with exams… what a relief!"
]

for text in examples:
    result = emotion_pipeline(text)
    sorted_result = sorted(result[0], key=lambda x: x['score'], reverse=True)
    print(f"\n🔹 Text: {text}")
    print(f"Top emotion → {sorted_result[0]['label']} ({sorted_result[0]['score']:.2f})")
```
